### 1. Pupularity Based Recommender System:

```python
books_ratings = rating.merge(books,on = 'ISBN')
books_ratings.head()
```

Number of Ratings on Each Book:

```python
num_of_ratings = books_ratings.groupby('Book-Title')['Book-Rating'].count().reset_index()
num_of_ratings.rename(columns={'Book-Rating':'Number of Ratings'},inplace=True)
num_of_ratings.head()
```

Average Ratings on Each Book:

```python
avg_of_ratings = books_ratings.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_of_ratings.rename(columns={'Book-Rating':'Average of Ratings'},inplace=True)
avg_of_ratings.head()
```

Merge Number of Ratings and Average Ratings DF

```python
popular_df = num_of_ratings.merge(avg_of_ratings, on = 'Book-Title')
popular_df.head()
```

So for Popularity Based Recommender System We will display the top 50 books on with atleast 250 ratings:

```python
popular_df = popular_df[popular_df['Number of Ratings'] >= 250].sort_values(by='Average of Ratings',ascending=False)
```

```python
popular_df.head()
```

Now we need the Authour Name, Publisher Name and Picture link So we will merge it with BOOKS dataset.

```python
popular_df.merge(books,on="Book-Title").drop_duplicates('Book-Title')
```

```python
final_popular_df = popular_df.merge(books,on="Book-Title").drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','Number of Ratings','Average of Ratings']]
```

```python
final_popular_df
```

So these are our most popular books with atleast 250+ Number of Ratings

```python
import pickle
pickle.dump(final_popular_df,open('popular.pkl','wb'))
```

### 2. Collaborative Filtering Based Recommender System

So in this type of recommendation system we will recommend the books on the base of the ratings from the users:
1. We will only consider those users that rated atleast 200 books and those books that got atleast ratings from 50 users.
2. Users will be our features and Books-Name will be our index

```python
books_ratings
```

Getting the users and how many times they rate the book(On Base of Rating):

```python
no_of_rating_users = books_ratings.groupby('User-ID')['Book-Rating'].count()>200
no_of_rating_users = no_of_rating_users[no_of_rating_users].index
```

```python
filtered_ratings = books_ratings[books_ratings['User-ID'].isin(no_of_rating_users)]
filtered_ratings.head()
```

Filtering on Base of Books with >50 ratings

```python
ratings_on_books = filtered_ratings.groupby('Book-Title')['Book-Rating'].count() >=50
ratings_on_books = ratings_on_books[ratings_on_books].index
```

```python
ratings_on_books
```

```python
filtered_ratings[filtered_ratings['Book-Title'].isin(ratings_on_books)]
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(ratings_on_books)]
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt
```

```python
pt.shape
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

Now we can represent each book with 810 users and each of then form a vector.<br>
Like book1 = [user1,user2,......,user810] so we will calculate the Eucledian distance b/w each vetcor and on base of that distance we will recommend the book we nearest distance.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
similarity_score
```

Now build a function and get suggestions of recommended books

```python
print(np.where(pt.index=='1984'))#feteching index
print(np.where(pt.index=='1984')[0][0])
```

```python
#Getting the top 5 books with max similarity(using enumerate to make index)
sorted(list(enumerate(similarity_score[0])),key=lambda x: x[1],reverse=True)[1:6]
```

```python
def recommend(book_name):
    #fetching index
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x: x[1],reverse=True)[1:4]
    data = []
    for i in similar_items:
        item = []
        #print(i[0]) # It will return the index of similar books
        #print(pt.index[i[0]]) # it will return the names of similar books
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    return data
```

```python
recommend('The Hobbit : The Enchanting Prelude to The Lord of the Rings')[0][2]
```
